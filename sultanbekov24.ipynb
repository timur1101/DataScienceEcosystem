```python
from datasets import Dataset

SYSTEM_PROMPT = """РўСӢ вҖ” СҚРәСҒРҝРөСҖСӮ РҝРҫ Familyguy. РһСӮРІРөСҮР°Р№ СӮРҫСҮРҪРҫ, \
РҪР° РҫСҒРҪРҫРІРө РёСҒСӮРҫСҖРёСҮРөСҒРәРёС… С„Р°РәСӮРҫРІ, РҪР°  Р°РҪРіР»РёР№СҒРәРҫРј СҸР·СӢРәРө. \
Р”Р°РІР°Р№ СҖР°Р·РІС‘СҖРҪСғСӮСӢРө РҫСӮРІРөСӮСӢ РёР· 2-5 РҝСҖРөРҙР»РҫР¶РөРҪРёР№."""

def format_to_chat(example, tokenizer):
    """РҹСҖРөРҫРұСҖР°Р·СғРөСӮ РҝР°СҖСғ question/answer РІ chat template С„РҫСҖРјР°СӮ Qwen3"""
    messages = [
        {"role": "system",    "content": SYSTEM_PROMPT},
        {"role": "user",      "content": example["question"]},
        {"role": "assistant", "content": example["answer"]}
    ]
    # enable_thinking=False вҖ” РҫСӮРәР»СҺСҮР°РөРј СҖРөР¶РёРј СҖР°Р·РјСӢСҲР»РөРҪРёСҸ РҝСҖРё РҫРұСғСҮРөРҪРёРё
    # РўР°Рә РјРҫРҙРөР»СҢ СғСҮРёСӮСҒСҸ СҒСҖР°Р·Сғ РҙР°РІР°СӮСҢ РҫСӮРІРөСӮ, РұРөР· <think>...</think> РұР»РҫРәР°
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=False,
        enable_thinking=False
    )
    return {"text": text}

print("вң… РӨСғРҪРәСҶРёСҸ С„РҫСҖРјР°СӮРёСҖРҫРІР°РҪРёСҸ РҫРҝСҖРөРҙРөР»РөРҪР°")
print(f"\nРЎРёСҒСӮРөРјРҪСӢР№ РҝСҖРҫРјРҝСӮ:")
print(f"  '{SYSTEM_PROMPT[:80]}...'")
```

```python
from unsloth import FastLanguageModel
import torch

# в”Җв”Җ РҹР°СҖР°РјРөСӮСҖСӢ Р·Р°РіСҖСғР·РәРё в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
MODEL_NAME     = "unsloth/Qwen3-4B"   # РұР°Р·РҫРІР°СҸ РјРҫРҙРөР»СҢ
MAX_SEQ_LENGTH = 2048                  # РјР°РәСҒРёРјР°Р»СҢРҪР°СҸ РҙР»РёРҪР° РәРҫРҪСӮРөРәСҒСӮР°
LOAD_IN_4BIT   = True                  # QLoRA: РәРІР°РҪСӮРёР·Р°СҶРёСҸ РІ 4-bit вҶ’ РјРөРҪСҢСҲРө VRAM

print(f"Р—Р°РіСҖСғР¶Р°РөРј РјРҫРҙРөР»СҢ: {MODEL_NAME}")
print(f"РҡРІР°РҪСӮРёР·Р°СҶРёСҸ: {'4-bit (QLoRA)' if LOAD_IN_4BIT else '16-bit (LoRA)'}")
print(f"РңР°РәСҒРёРјР°Р»СҢРҪСӢР№ РәРҫРҪСӮРөРәСҒСӮ: {MAX_SEQ_LENGTH} СӮРҫРәРөРҪРҫРІ")
print("\nвҸі Р—Р°РіСҖСғР¶Р°РөРј... (3-5 РјРёРҪСғСӮ)")

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name      = MODEL_NAME,
    max_seq_length  = MAX_SEQ_LENGTH,
    load_in_4bit    = LOAD_IN_4BIT,
    dtype           = None,            # auto: bfloat16 РҪР° T4
)

print("\nвң… Р‘Р°Р·РҫРІР°СҸ РјРҫРҙРөР»СҢ Р·Р°РіСҖСғР¶РөРҪР°!")
# РЎРәРҫР»СҢРәРҫ VRAM Р·Р°РҪРёРјР°РөСӮ РұР°Р·РҫРІР°СҸ РјРҫРҙРөР»СҢ
vram_used = torch.cuda.memory_allocated() / 1024**3
print(f"   VRAM РёСҒРҝРҫР»СҢР·РҫРІР°РҪРҫ: {vram_used:.1f} GB")
```

```python
# в”Җв”Җ Р”РҫРұР°РІР»СҸРөРј LoRA-Р°РҙР°РҝСӮРөСҖСӢ в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
#
# LoRA РҙРҫРұР°РІР»СҸРөСӮ РјР°Р»РөРҪСҢРәРёРө РјР°СӮСҖРёСҶСӢ (rank=16) Рә РәР»СҺСҮРөРІСӢРј СҒР»РҫСҸРј СӮСҖР°РҪСҒС„РҫСҖРјРөСҖР°.
# РһРұСғСҮР°СҺСӮСҒСҸ СӮРҫР»СҢРәРҫ СҚСӮРё РјР°СӮСҖРёСҶСӢ вҖ” РҫСҒСӮР°Р»СҢРҪСӢРө РІРөСҒР° Р·Р°РјРҫСҖРҫР¶РөРҪСӢ.
# РӯСӮРҫ РҝРҫР·РІРҫР»СҸРөСӮ РҫРұСғСҮРёСӮСҢ РјРҫРҙРөР»СҢ РҪР° T4 Р·Р° СҖР°Р·СғРјРҪРҫРө РІСҖРөРјСҸ!

model = FastLanguageModel.get_peft_model(
    model,
    r                  = 16,           # СҖР°РҪРі LoRA: 8=РұСӢСҒСӮСҖРөРө, 32=СғРјРҪРөРө
    lora_alpha         = 16,           # РјР°СҒСҲСӮР°Рұ LoRA (РҫРұСӢСҮРҪРҫ = r)
    lora_dropout       = 0,            # dropout (0 = СҒСӮР°РұРёР»СҢРҪРөРө)
    target_modules     = [             # Рә РәР°РәРёРј СҒР»РҫСҸРј РҙРҫРұР°РІР»СҸРөРј Р°РҙР°РҝСӮРөСҖСӢ
        "q_proj", "k_proj", "v_proj", "o_proj",   # attention
        "gate_proj", "up_proj", "down_proj"        # feed-forward
    ],
    bias               = "none",
    use_gradient_checkpointing = "unsloth",  # СҚРәРҫРҪРҫРјРёСӮ VRAM РҙР»СҸ РҙР»РёРҪРҪСӢС… РәРҫРҪСӮРөРәСҒСӮРҫРІ
    random_state       = 42,
)

# РЎСҮРёСӮР°РөРј РҫРұСғСҮР°РөРјСӢРө РҝР°СҖР°РјРөСӮСҖСӢ
trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
total     = sum(p.numel() for p in model.parameters())
print(f"вң… LoRA-Р°РҙР°РҝСӮРөСҖСӢ РҙРҫРұР°РІР»РөРҪСӢ!")
print(f"   РһРұСғСҮР°РөРјСӢС… РҝР°СҖР°РјРөСӮСҖРҫРІ: {trainable:,} ({100*trainable/total:.2f}% РҫСӮ РІСҒРөС…)")
print(f"   Р—Р°РјРҫСҖРҫР¶РөРҪРҫ РҝР°СҖР°РјРөСӮСҖРҫРІ: {total-trainable:,}")
print(f"\nрҹ’Ў РңСӢ РҫРұСғСҮР°РөРј СӮРҫР»СҢРәРҫ {100*trainable/total:.2f}% РјРҫРҙРөР»Рё вҖ” СҚСӮРҫ Рё РөСҒСӮСҢ РјР°РіРёСҸ LoRA!")
```

```python
from unsloth.chat_templates import get_chat_template
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
# Р’РәР»СҺСҮР°РөРј СҖРөР¶РёРј РёРҪС„РөСҖРөРҪСҒР° (РҫСӮРәР»СҺСҮР°РөСӮ dropout Рё СӮ.Рҙ.)
FastLanguageModel.for_inference(model)

def ask_model(question, system=SYSTEM_PROMPT, max_new_tokens=300, thinking=False):
    """Р—Р°РҙР°С‘Рј РІРҫРҝСҖРҫСҒ РјРҫРҙРөР»Рё Рё РҝРҫР»СғСҮР°РөРј РҫСӮРІРөСӮ"""
    messages = [
        {"role": "system",    "content": system},
        {"role": "user",      "content": question}
    ]
    inputs = tokenizer.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        enable_thinking=thinking,    # True = РҙСғРјР°РөСӮ РІСҒР»СғС…, False = СҒСҖР°Р·Сғ РҫСӮРІРөСҮР°РөСӮ
        return_tensors="pt"
    ).to("cuda")

    with torch.no_grad():
        outputs = model.generate(
            input_ids      = inputs,
            max_new_tokens = max_new_tokens,
            temperature    = 0.7,
            top_p          = 0.8,
            top_k          = 20,
            do_sample      = True,
            pad_token_id   = tokenizer.eos_token_id
        )

    # Р”РөРәРҫРҙРёСҖСғРөРј СӮРҫР»СҢРәРҫ РҪРҫРІСӢРө СӮРҫРәРөРҪСӢ (РұРөР· РҝСҖРҫРјРҝСӮР°)
    new_tokens = outputs[0][inputs.shape[-1]:]
    response   = tokenizer.decode(new_tokens, skip_special_tokens=True)

    # РЈРұРёСҖР°РөРј <think>...</think> РұР»РҫРә РөСҒР»Рё РҫРҪ РөСҒСӮСҢ
    if "</think>" in response:
        response = response.split("</think>")[-1].strip()

    return response


# РўРөСҒСӮРҫРІСӢРө РІРҫРҝСҖРҫСҒСӢ РҙР»СҸ СҒСҖР°РІРҪРөРҪРёСҸ РҙРҫ/РҝРҫСҒР»Рө
TEST_QUESTIONS = [
    "How is Meg treated by the family?",
    "How does Brian differ from typical pets?",
    "What makes Family GuyвҖҷs humor unique?",
    "How does Family Guy generate revenue?"

]

print("=" * 60)
print("РһРўР’Р•РўР« Р‘РҗР—РһР’РһРҷ РңРһР”Р•РӣРҳ (Р”Рһ РһР‘РЈР§Р•РқРҳРҜ)")
print("=" * 60)

baseline_answers = {}
for q in TEST_QUESTIONS:
    print(f"\nвқ“ {q}")
    answer = ask_model(q)
    baseline_answers[q] = answer
    print(f"рҹӨ– {answer}")
    print("-" * 40)
```

```python
from trl import SFTTrainer, SFTConfig

trainer = SFTTrainer(
    model        = model,
    tokenizer    = tokenizer,
    train_dataset= formatted_dataset,
    args = SFTConfig(
        # в”Җв”Җ РһСҒРҪРҫРІРҪСӢРө РҝР°СҖР°РјРөСӮСҖСӢ в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
        output_dir                  = "./qwen3_kz_history",
        num_train_epochs            = 20,     # вҶ‘ РұСӢР»Рҫ 3 вҖ” РұРҫР»СҢСҲРө СҚРҝРҫС… РҙР»СҸ Р·Р°РҝРҫРјРёРҪР°РҪРёСҸ С„Р°РәСӮРҫРІ
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,      # СҚС„С„РөРәСӮРёРІРҪСӢР№ batch_size = 2*4 = 8
        max_seq_length              = MAX_SEQ_LENGTH,

        # в”Җв”Җ РһРҝСӮРёРјРёР·Р°СӮРҫСҖ в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
        learning_rate               = 5e-5,   # вҶ“ РұСӢР»Рҫ 2e-4 вҖ” РјРөРҪСҢСҲРө lr РҝСҖРё РұРҫР»СҢСҲРёС… СҚРҝРҫС…Р°С…
        lr_scheduler_type           = "cosine", # РҝР»Р°РІРҪРҫРө СҒРҪРёР¶РөРҪРёРө lr
        warmup_ratio                = 0.1,    # вҶ‘ РұСӢР»Рҫ 0.05 вҖ” РұРҫР»СҢСҲРө СҖР°Р·РҫРіСҖРөРІ РҝСҖРё РҙРҫР»РіРҫРј РҫРұСғСҮРөРҪРёРё
        optim                       = "adamw_8bit", # 8-bit Adam вҶ’ РјРөРҪСҢСҲРө VRAM
        weight_decay                = 0.01,

        # в”Җв”Җ РӣРҫРіРёСҖРҫРІР°РҪРёРө в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
        logging_steps               = 10,       # РІСӢРІРҫРҙРёСӮСҢ loss РәР°Р¶РҙСӢРө 10 СҲР°РіРҫРІ
        save_steps                  = 100,      # СҒРҫС…СҖР°РҪСҸСӮСҢ checkpoint РәР°Р¶РҙСӢРө 100 СҲР°РіРҫРІ
        save_total_limit            = 2,        # С…СҖР°РҪРёСӮСҢ СӮРҫР»СҢРәРҫ 2 РҝРҫСҒР»РөРҙРҪРёС… checkpoint

        # в”Җв”Җ РўРҫСҮРҪРҫСҒСӮСҢ в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
        bf16                        = torch.cuda.is_bf16_supported(),
        fp16                        = not torch.cuda.is_bf16_supported(),

        # в”Җв”Җ РҹСҖРҫСҮРөРө в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
        dataset_text_field          = "text",
        seed                        = 42,
        report_to                   = "none",   # РҫСӮРәР»СҺСҮР°РөРј wandb/tensorboard
    )
)

# РҹРҫРәР°Р·СӢРІР°РөРј СҒСӮР°СӮРёСҒСӮРёРәСғ РҝР°РјСҸСӮРё РҙРҫ РҫРұСғСҮРөРҪРёСҸ
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024**3, 3)
max_memory = round(gpu_stats.total_memory / 1024**3, 3)
print(f"VRAM Р·Р°СҖРөР·РөСҖРІРёСҖРҫРІР°РҪРҫ РҙРҫ РҫРұСғСҮРөРҪРёСҸ: {start_gpu_memory} GB")
print(f"Р’СҒРөРіРҫ VRAM РҙРҫСҒСӮСғРҝРҪРҫ: {max_memory} GB")
print(f"\nрҹҡҖ РқР°СҮРёРҪР°РөРј РҫРұСғСҮРөРҪРёРө... (~28-30 РјРёРҪ РҙР»СҸ ~400 РҝР°СҖ, 10 СҚРҝРҫС…)")
print("   РЎР»РөРҙРё Р·Р° loss вҖ” РҫРҪ РҙРҫР»Р¶РөРҪ СғРұСӢРІР°СӮСҢ!")

import time
start_time = time.time()

trainer_stats = trainer.train()

elapsed = time.time() - start_time
print(f"\nвң… РһРұСғСҮРөРҪРёРө Р·Р°РІРөСҖСҲРөРҪРҫ!")
print(f"   Р’СҖРөРјСҸ: {elapsed/60:.1f} РјРёРҪСғСӮ")
print(f"   РӨРёРҪР°Р»СҢРҪСӢР№ loss: {trainer_stats.training_loss:.4f}")
used_memory = round(torch.cuda.max_memory_reserved() / 1024**3, 3)
print(f"   РңР°РәСҒРёРјР°Р»СҢРҪСӢР№ СҖР°СҒС…РҫРҙ VRAM: {used_memory} GB РёР· {max_memory} GB")
```
